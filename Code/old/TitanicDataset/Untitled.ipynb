```python
Y=dftrain['target']
```

```python
dftrain.drop('target',inplace=True,axis=1)
```

```python
dftrain.drop("id",inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(dftrain, Y, test_size=0.00000001,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#StandardScaler
sc2=StandardScaler()
sc2.fit(X_train)
X_train2=sc2.transform(X_train)
X_test2=sc2.transform(X_test)
Y_train2=Y_train
Y_test2=Y_test
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rn=RandomForestClassifier(max_depth=2, random_state=0)
```

```python
 rn.fit(X_train2,Y_train2)
```

```python
 sgc=SGDClassifier(max_iter=100000, tol=1e-5)
```

```python
sgc.fit(X_train2,Y_train2)
```

```python
sgc.score(X_test2,Y_test2)
```
